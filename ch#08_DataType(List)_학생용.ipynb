## Python Collections (Arrays)
There are four collection data types in the Python programming language:

* **List** is a collection which is <u>ordered and changeable. Allows duplicate members.</u>
* **Tuple** is a collection which is <u>ordered and unchangeable. Allows duplicate members.</u>
* **Set** is a collection which is <u>unordered and unindexed. No duplicate members.</u>
* **Dictionary** is a collection which is <u>unordered, changeable and indexed. No duplicate members.</u>

When choosing a collection type, it is useful to understand the properties of that type. Choosing the right type for a particular data set could mean retention of meaning, and, it could mean an increase in efficiency or security.

# 4. <font color=red>리스트 메소드(Very Very Important)</font>
-리스트를 활용한다는 것은 리스트의 메소드를 사용한다는것과 동일.  
-메소드를 얼마나 잘 사용하느냐가 리스트의 실력, 나아가 파이썬의 실력을 결정한다.  
-모두 11개의 메소드밖에 없음. 10개가 매우 쉬움  

* **method 정리**  
a.append(x) # a 끝에 x 추가 (a+=[x]와 동일)  
a.extend([1,2]) # a 끝에 [1,2]의 모든 원소 추가 (a+=[1,2]와 동일)  
a.insert(i, x) # a의 인덱스 i 자리에 x 끼워넣기 (a[i:i]=[x]와 동일)  
a.remove(x) # a에서 첫번째 나오는 x 제거  
a.pop() # a의 마지막 원소 제거 후 반환  
a.pop(i) # a[i] 제거 후 반환  
a.clear() # a=[]  
a.index(x) # a에서 원소가 x인 첫 인덱스 반환  
a.count(x) # a에서 x가 몇 개 있는지 반환  
a.sort() # a를 오름차순으로 정렬  
a.sort(reverse=True) # a를 내림차순으로 정렬  
a.reverse() # a의 원소를 역순으로 재배치  
b=a.copy() # b=a[:]  

* append, clear, copy 는 수업시간에 설명  

* count,extend,index,insert,pop,remove,reverse,sort 중에 임의로 5개 선정 후 수행평가  
형식은 1학기 built-in 함수 수행평가와 동일함.
